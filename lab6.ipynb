```python
def line_params(x, y):
    x = np.array(x);
    y = np.array(y);
    m = ( ( ( np.mean(x)*np.mean(y) ) - np.mean(x*y) ) /
          ( ( np.mean(x)*np.mean(x) ) - np.mean(x*x) ) )
    m=round(m,2)
    c=(np.mean(y) - m*np.mean(x))
    c=round(c,2)
    
    return m, c
```
